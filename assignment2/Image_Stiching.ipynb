```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
R=cv2.imread("lab1.jpeg")
L=cv2.imread("lab2.jpeg")
R=R[:,:,::-1]
L=L[:,:,::-1]
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(R)
plt.subplot(122)
plt.imshow(L)
```

```python
(status,stitched)=cv2.Stitcher_create().stitch([R,L])
```

```python
plt.figure(figsize=(10,10))
if status ==0:
    plt.imshow(stitched)
else:
    print("[INFO] image stitching failed ({})".format(status))
```
